```python
'''
This code updates team names in the 'ipl_match_df' DataFrame, replacing 'Delhi Daredevils' with 'Delhi Capitals' and 'Deccan Chargers' with 'Sunrisers Hyderabad.' 
It then filters the DataFrame to include only rows where 'team1' and 'team2' values are within the 'ipl_teams' list. 
The shape of the resulting DataFrame is notated.
'''

ipl_teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]

ipl_match_df['team1'] = ipl_match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
ipl_match_df['team2'] = ipl_match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
ipl_match_df['team1'] = ipl_match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
ipl_match_df['team2'] = ipl_match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

ipl_match_df = ipl_match_df[ipl_match_df['team1'].isin(ipl_teams)]
ipl_match_df = ipl_match_df[ipl_match_df['team2'].isin(ipl_teams)]

ipl_match_df.shape
```

```python
'''
This code filters the 'ipl_match_df' DataFrame to include only rows where 'dl_applied' is equal to 0. 
Then, it selects specific columns ('match_id,' 'city,' 'winner,' and 'total_runs') and stores the result in 'ipl_match_df.' 
Finally, it merges 'ipl_match_df' with another DataFrame called 'deliver' on the 'match_id' column and retains only rows where 'inning' is equal to 2, storing the result in 'deliver_df.'
'''

ipl_match_df = ipl_match_df[ipl_match_df['dl_applied'] == 0]
ipl_match_df = ipl_match_df[['match_id','city','winner','total_runs']]
deliver_df = ipl_match_df.merge(deliver,on='match_id')
deliver_df = deliver_df[deliver_df['inning'] == 2]

deliver_df
```

```python
'''
This code calculates and adds three new columns to the 'deliver_df' DataFrame:

'current_score': It calculates the cumulative sum of 'total_runs_y' grouped by 'match_id' and stores it in the 'current_score' column.
'runs_left': It calculates the runs remaining to be chased by subtracting 'current_score' from 'total_runs_x'.
'balls_left': It calculates the number of balls remaining in the innings by subtracting the cumulative ball count (over * 6 + ball) from the total of 126.
'''

deliver_df['current_score'] = deliver_df.groupby('match_id').cumsum()['total_runs_y']
deliver_df['runs_left'] = deliver_df['total_runs_x'] - deliver_df['current_score']
deliver_df['balls_left'] = 126 - (deliver_df['over']*6 + deliver_df['ball'])

deliver_df
```

```python
'''
This code preprocesses the 'player_dismissed' column in the DataFrame 'deliver_df'. It fills missing values with '0', converts non-zero values to '1', and then changes the data type to integers. 
It calculates the cumulative sum of 'player_dismissed' grouped by 'match_id' to derive the number of wickets taken in each match, and this information is stored in a new 'wickets' column.
'''

deliver_df['player_dismissed'] = deliver_df['player_dismissed'].fillna("0")
deliver_df['player_dismissed'] = deliver_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
deliver_df['player_dismissed'] = deliver_df['player_dismissed'].astype('int')
wickets = deliver_df.groupby('match_id').cumsum()['player_dismissed'].values
deliver_df['wickets'] = 10 - wickets

#displays above 5 records
deliver_df.head()
```

```python
#displays below 5 records
deliver_df.tail()
```

```python
'''
Calculates the current run rate (crr) and required run rate (rrr) for each row in the DataFrame `deliver_df`. 
Additionally, it defines a function `result` to determine the match result (1 for win, 0 for loss) and adds a new column 'result' based on the match outcome. 
Finally, it creates a new DataFrame `ipl_final_df` by selecting specific columns from `deliver_df`, shuffles the rows, and retrieves a random sample.
'''

deliver_df['crr'] = (deliver_df['current_score']*6)/(120 - deliver_df['balls_left'])
deliver_df['rrr'] = (deliver_df['runs_left']*6)/deliver_df['balls_left']

def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0

deliver_df['result'] = deliver_df.apply(result,axis=1)

ipl_final_df = deliver_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
ipl_final_df = ipl_final_df.sample(ipl_final_df.shape[0])
ipl_final_df.sample()
```

```python
'''
Preprocesses the DataFrame `ipl_final_df` by removing rows with missing values and rows where 'balls_left' equals 0. 
Then, it separates the features (x) and the target variable (y) for further analysis.
'''

ipl_final_df.dropna(inplace=True)
ipl_final_df = ipl_final_df[ipl_final_df['balls_left'] != 0]
x = ipl_final_df.iloc[:,:-1]
y = ipl_final_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting the data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#display
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
#Creates a scikit-learn `ColumnTransformer` named `trf` that applies `OneHotEncoder` to specific columns ('batting_team', 'bowling_team', 'city') and passes through the remaining columns.
trf = ColumnTransformer([('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
#Creates a scikit-learn pipeline named `pipe` with two steps: `trf` and `LogisticRegression` using the 'liblinear' solver.
pipe = Pipeline(steps=[('step1',trf),('step2',LogisticRegression(solver='liblinear'))])
```

```python
#Fits the model pipeline `pipe` on the training data `x_train` and `y_train`.
pipe.fit(x_train,y_train)
```

```python
#Generates predicted values y_pred using the model pipeline 'pipe' on the test data x_test.
y_pred = pipe.predict(x_test)
```

```python
#Calculates the accuracy score by comparing the predicted values y_pred with the actual values y_test.
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
#Predicts the probability using the model for the 10th item in the test data.
pipe.predict_proba(x_test)[10]
```

```python
'''
Code has two functions:
 match_summary:  prints a summary of teams and target runs.
 match_progression: calculates match stats and returns a DataFrame, also printing the target runs.
'''

def match_summary(row):
    print("Batting Team-" + row['batting_team'] + " | Bowling Team-" + row['bowling_team'] + " | Target- " + str(row['total_runs_x']))

def match_progression(x_df,match_id,pipe):
    ipl_match = x_df[x_df['match_id'] == match_id]
    ipl_match = ipl_match[(ipl_match['ball'] == 6)]
    temp_df = ipl_match[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
    
temp_df,target = match_progression(deliver_df,74,pipe)
temp_df    
```

```python
cities = ['Hyderabad', 'Bangalore', 'Mumbai', 'Indore', 'Kolkata', 'Delhi',
       'Chandigarh', 'Jaipur', 'Chennai', 'Cape Town', 'Port Elizabeth',
       'Durban', 'Centurion', 'East London', 'Johannesburg', 'Kimberley',
       'Bloemfontein', 'Ahmedabad', 'Cuttack', 'Nagpur', 'Dharamsala',
       'Visakhapatnam', 'Pune', 'Raipur', 'Ranchi', 'Abu Dhabi',
       'Sharjah', 'Mohali', 'Bengaluru']
```

```python
# Load your pickled model (assuming it's a scikit-learn pipeline)
# Define Streamlit app title
# Create two columns for team selection
# Create select boxes for team and city selection
# Input target score
# Create three columns for score, overs, and wickets input
# Calculate various match parameters
# Prepare input data as a DataFrame
# Make model predictions
# Display win and loss probabilities

pipe = pickle.load(open('pipe.pkl', 'rb'))

st.title('IPL Win Predictor')

# Create columns using st.columns
col1, col2 = st.columns(2)

with col1:
    batting_team = st.selectbox('Select the batting team', sorted(ipl_teams))

with col2:
    bowling_team = st.selectbox('Select the bowling team', sorted(ipl_teams))

selected_city = st.selectbox('Select host city', sorted(cities))

target = st.number_input('Target')

# Create columns using st.columns
col3, col4, col5 = st.columns(3)

with col3:
    score = st.number_input('Score')

with col4:
    overs = st.number_input('Overs completed')

with col5:
    wickets = st.number_input('Wickets out')

if st.button('Predict Probability'):
    runs_left = target - score
    balls_left = 120 - (overs * 6)
    wickets = 10 - wickets
    crr = score / overs
    rrr = (runs_left * 6) / balls_left

    input_df = pd.DataFrame({
        'batting_team': [batting_team],
        'bowling_team': [bowling_team],
        'city': [selected_city],
        'runs_left': [runs_left],
        'balls_left': [balls_left],
        'wickets': [wickets],
        'total_runs_x': [target],
        'crr': [crr],
        'rrr': [rrr]
    })

    result = pipe.predict_proba(input_df)
    loss = result[0][0]
    win = result[0][1]

    st.header(batting_team + "- " + str(round(win * 100)) + "%")
    st.header(bowling_team + "- " + str(round(loss * 100)) + "%")
```
